# Regression Test for Beat Detection

```python
import librosa
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats
```

```python
# load the audio track
y, sr = librosa.load("mixed-grooves-straight_91bpm.wav")
tempo, beats = librosa.beat.beat_track(y=y, sr=sr)
beats = librosa.frames_to_time(beats)
x = np.arange(beats.shape[0])
plt.xlabel("Beat Number")
plt.ylabel("Time")

# define the linear regression
slope, intercept, r, p, std_err = stats.linregress(x, beats)

def myfunc(x):
    return slope * x + intercept

model = list(map(myfunc, x))
plt.scatter(x, beats)
plt.plot(x, model, color='purple')
plt.show()
```
